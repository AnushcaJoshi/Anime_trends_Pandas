### 2. Top 5 genres that Anime lovers prefer the most

```python
sorted_df1 = df.sort_values(by='Favorites', ascending=False)

top_5_favorites = sorted_df1.head(5)

# Initializing an empty dictionary to store the genre counts
genre_counts = {}

for index, row in top_5_favorites.iterrows():
    genres = [genre.strip() for genre in row['Genres'].split(',')]
    for genre in genres:
        genre_counts[genre] = genre_counts.get(genre, 0) + 1

sorted_genre_counts = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)

top_5_genres = dict(sorted_genre_counts[:5])

print("Anime names with their counts from top 5 favourite Anime List", top_5_genres)
```

### 3. Where do People watch most animes: TV, Mobile or OVA?

### 4. Is there any correlation between dropping to watch Animes with episodes, source, duration, and members

```python
# Create subplots for each correlation plot
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plot correlation between 'Dropped' and 'Episodes'
axes[0, 0].scatter(df['Episodes'], df['Dropped'])
axes[0, 0].set_xlabel('Episodes')
axes[0, 0].set_ylabel('Dropped')

# Plot correlation between 'Dropped' Episodesand 'Source'
axes[0, 1].scatter(df['Source'], df['Dropped'])
axes[0, 1].set_xlabel('Source')
axes[0, 1].set_ylabel('Dropped')

# Plot correlation between 'Dropped' and 'Duration'
axes[1, 0].scatter(df['Duration'], df['Dropped'])
axes[1, 0].set_xlabel('Duration')
axes[1, 0].set_ylabel('Dropped')

# Plot correlation between 'Dropped' and 'Members'
axes[1, 1].scatter(df['Members'], df['Dropped'])
axes[1, 1].set_xlabel('Members')
axes[1, 1].set_ylabel('Dropped')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
source_counts = df['Source'].value_counts()

plt.figure(figsize=(10, 6))

sns.barplot(x=source_counts.index[:5], y=source_counts.values[:5])
plt.title('Source Counts (Top 5)')
plt.xlabel('Source')
plt.ylabel('Dropped')

plt.xticks(range(len(source_counts.index[:5])), ['Original', 'Manga', 'Novel'])

plt.show()
```

#### Anime drop rate is higher in Original Animes as compared to all other sources
